```python
Corpus=pd.read_csv('/content/gdrive/MyDrive/NCKH/Data/QA_number_label.csv')
Corpus.head()
len(Corpus)
sns.countplot(Corpus.Intent)
plt.xlabel("Intent")
plt.title("countplot")
X_train,X_test,y_train,y_test=model_selection.train_test_split(Corpus['Question'],Corpus['Intent'],test_size=0.2)
Encoder=LabelEncoder()
Test_Y=Encoder.fit_transform(y_test)
Train_Y=Encoder.fit_transform(y_train)
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf_vect=TfidfVectorizer(max_features=400)
Tfidf_vect.fit(Corpus['Question'])
Train_X_Tfidf=Tfidf_vect.transform(X_train)
Test_X_Tfidf=Tfidf_vect.transform(X_test)
```

```python
print(Train_X_Tfidf)
```

```python
#@title  NAIVE BAYES

# #Predict the response for test dataset
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import ComplementNB
from sklearn.model_selection import GridSearchCV

import time
start_time = time.time()
param_grid = {
    "alpha": [0.1],
}
Naive = GridSearchCV(
    MultinomialNB(), param_grid,
)
Naive.fit(Train_X_Tfidf,y_train)
predictions_NB=Naive.predict(Test_X_Tfidf)
print("Best score")
print (Naive.best_score_) 
print(Naive.best_estimator_) 
grid_predictions = Naive.predict(Test_X_Tfidf) 
print(classification_report(y_test, grid_predictions)) 

# print("Accuracy:",metrics.accuracy_score(y_test, predictions_NB))
# Naive=naive_bayes.MultinomialNB()
# Naive.fit(Train_X_Tfidf,y_train)
# predictions_NB=Naive.predict(Test_X_Tfidf)
# print('Best score:', Naive.best_score_*100)
# print('Best C:',Naive.best_estimator_.alpha)
print("Accuracy:",metrics.accuracy_score(y_test, predictions_NB)*100)
# print('Best score:', Naive.best_score_*100)
# print(Naive.best_params_) 
end_time = time.time()
elapsed_time = end_time - start_time
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")

# print(classification_report(y_test,predictions_NB))
```
