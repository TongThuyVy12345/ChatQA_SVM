```python
X_train,X_test,y_train,y_test=train_test_split(Corpus['Question'],Corpus['Intent'],test_size=0.1, random_state=32)
```

```python
Test_Y=y_test
Train_Y=y_train
Train_Y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf_vect=TfidfVectorizer(max_features=400)
Tfidf_vect.fit(Corpus['Question'])
Train_X_Tfidf=Tfidf_vect.transform(X_train)
Test_X_Tfidf=Tfidf_vect.transform(X_test)

```
