```python
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the dataset
docs_to_train = ["Do My", 
                 "Uyen Do", 
                 "Tong Vy", 
                 "Do Thu"]
y_train = [4, 0, 3, 2] # 1 for positive, 0 for negative, 2 for neutral

# Convert the text data into numerical data using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X_train = vectorizer.fit_transform(docs_to_train)

# Train a linear SVM
clf = SVC(kernel='linear', C=1, random_state=0)
clf.fit(X_train, y_train)

# Predict on a new document
docs_to_predict = [input("Enter a document to predict: ")]
X_predict = vectorizer.transform(docs_to_predict)
y_predict = clf.predict(X_predict)
print("Predicted class:", y_predict[0])
```

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the dataset
docs_to_train = ["Do My", 
                 "Uyen Do", 
                 "Tong Vy", 
                 "Do Thu"]
y_train = [4, 0, 3, 2] # 1 for positive, 0 for negative, 2 for neutral

# Convert the text data into numerical data using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
X_train = vectorizer.fit_transform(docs_to_train)

# Train a linear SVM
clf = SVC(kernel='linear', C=1, random_state=0)
clf.fit(X_train, y_train)

# Predict on new documents
while True:
    doc_to_predict = input("Question('Stop' to quit): ")
    if doc_to_predict == 'Stop':
        break
    X_predict = vectorizer.transform([doc_to_predict])
    y_predict = clf.predict(X_predict)
    print("Chatbot:", y_predict[0])
```
